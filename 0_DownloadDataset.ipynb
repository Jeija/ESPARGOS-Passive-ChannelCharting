### Optional: Plot Clustered Position Labels

```python
import cluster_utils
import CCEvaluation
import numpy as np

training_positions = []
for dataset in training_set_robot:
    cluster_utils.cluster_dataset(dataset)
    training_positions.append(dataset["cluster_positions"][:,:2])
training_positions = np.concatenate(training_positions)
CCEvaluation.plot_colorized(training_positions, training_positions, title = "Position Labels - Training Set (Robot)")

for dataset in test_set_robot:
    cluster_utils.cluster_dataset(dataset)
    CCEvaluation.plot_colorized(dataset["cluster_positions"][:,:2], dataset["cluster_positions"][:,:2], title = "Position Labels - Test Set (Robot)")

for dataset in test_set_human:
    cluster_utils.cluster_dataset(dataset)
    CCEvaluation.plot_colorized(dataset["cluster_positions"][:,:2], dataset["cluster_positions"][:,:2], title = "Position Labels - Test Set (Human)")
```
